# Assumptions
If the goal is **inference**, assumptions need to be checked rigorously. If the goal is **prediction**, some of assumptions can be relaxed. 
1. Linearity: a linear relationship exists between the target and residuals
2. Independence of errors: the residuals are not correlated
3. Normality of errors: the residuals are normally distributed
4. No perfect multicollinearity: the features are not perfectly correlated with each other
5. Equal variance of errors: the spread of residuals is consistent across predictions
