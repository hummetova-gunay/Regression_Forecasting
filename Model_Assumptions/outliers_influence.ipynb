# Outliers and Influential Data Points
These need to be considered during evaluation

There are two types of outliers: **Non-influential** and **Influential** 

**Non-influential** outliers do not have impact on the slope, the feature target relationship is maintained

**Influential**  outlier changes the fundamental relationship between x and y.

Note that dataset size matters a lot when it comes to be influenced by outliers,large datasets are impacted less

# Diagnosis : Cook's Distance
This measures the influence of a data point on a regression line. It can take a long time for calculation: we remove every single data point in our dataset and calculate the impact that removing each data point has on our model. If we remove a data point and slope does not change significantly then it is not influential. Cook's distance>1 is considered as significant problem. Cook's Distance> 0.5 worth investigating

```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import numpy as np
```

```python
diamonds = pd.read_csv('../Course Materials/Data/Diamonds Prices2022.csv')
```

```python
diamonds['carat_sq'] = diamonds['carat']**2
```

```python
diamonds =  diamonds.sample(300)
```

```python
features =[
    'carat',
    'carat_sq', 
    'depth',
    'table',
    'x'
]
X = sm.add_constant(diamonds.loc[:, features])
y = diamonds['price']

model = sm.OLS(y, X).fit()
```
