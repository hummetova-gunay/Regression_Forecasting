# Assumption: Features are not perfectly correlated with each other
This would lead to unreliable and illogical model coefficients. 

If two features have a correlation (r) of 1, there are infinite ways to minimize squared error. Conceptually these features are telling us the same information and model does not know how to assign value to each of these features.


String multicollinearity (r>0.7) can still cause big problems



# Diagnosis : Variance Inflation Factor (VIF)
- Each feature is treated as target and $R^2$ measures how well the other features predict it.

- rule of thumb: VIF>5 indicates that a variable is causing multicollinearity
