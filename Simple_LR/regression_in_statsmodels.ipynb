```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('../Course Materials/Data/Diamonds Prices2022.csv')
df.head()
```

```python
X = df[['carat']]
y =df['price']
```

```python
model_sm = sm.OLS(y, X).fit()
model_sm.summary()
```

Here we see R_squared  uncentered because we did not add a constant. `add_constant` adds one value to our dataframe. When we do not have that constant `Statsmodels` assumes we want our model to run through the origin (0,0). `sm.add_constant()` let statsmodels calculate  a y-intercept other than 0. On the other hand `sklearn` takes care of this itself

```python
X = sm.add_constant(df[['carat']])
y =df['price']
model_sm = sm.OLS(y, X).fit()
model_sm.summary()
```

### Interpreting the model
`coef` is used to interpret the output: `const` is intercept, `carat` is slope
