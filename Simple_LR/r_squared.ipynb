# Notes: $R^2$
$R^2$ is another very important measure of model accuracy which is also known as the coefficient of determination measures. How much better the model is at predicting the target than using its mean. 

When we do not use any features for prediction mean is our best guess. 

$R^2$ values are going to be bounded between 0 and 1. 0 would indicate that it is equivalent to our mean, whereas $r^2=1$ would indicate that we have explained 100% of the variance in our data around its mean using our model

$
R^2= 1 - \frac{SSE}{SST}
$

**SSE** - Sum of Squared error: distance between values and line

**SST** - Sum of Squared total: distance between values and mean
