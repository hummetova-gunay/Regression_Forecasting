# Statsmodels and Scikit-Learn: Preview
`Statsmodels` is ideal for inference but has some drawbacks like difficulty with leverage production and does not have latest and greatest algorithms  unlike `Scikit-Learn`

```python
import pandas as pd
import numpy as np
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../Course Materials/Data/Diamonds Prices2022.csv')
df.head()
```

```python
X = df[['carat']]
y = df['price']
```

```python
model_sm = sm.OLS(y,X).fit()
model_sm.summary()
```

```python
model_lr = LinearRegression(fit_intercept=False).fit(X,y)
model_lr.coef_
```
