# Hypothesis testing
Regression models include several hypothesis tests including F-test. It tells us whether our model is better at predicting our target than using the mean.
We are basically testing if our model is adding any value or it is completely useless.

**General steps**
- State the null and alternative hypothesis
- Set a significance level ($\alpha$) : cutoff point at which we make our decision. A lower ($\alpha$) means it is going to be hard to reject the null hypothesis. A higher ($\alpha$) means it will be easy to reject the null hypothesis. 
- Calculate test statistics and p-value: if p-value is less than or equal to ($\alpha$) then we reject the null hypothesis in favor of alternative hypothesis. (The model is not useless)


**F-statistics** - ratio in statistics that compares the variance between groups(explained variance) to the variance within groups (unexplained variance). a large F - value (>>1) means the model is useful (reject null), small F value means reject null. When we say `.summary` we can see F-statistics and Prob(F-statistics). They will help us decide whether we should make changes in our model or not. We use F-stastics to calculate p-value which is easier to interpret. 

To conclude our statistical test we are going to compare P value to our alpha.
