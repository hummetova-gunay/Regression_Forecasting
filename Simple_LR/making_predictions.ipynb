`.predict` is frequently used to make predictions for  a single datapoint or an entire dataframe

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('../Course Materials/Data/Diamonds Prices2022.csv')
df.head()
```

```python
X = sm.add_constant(df[['carat']])
y =df['price']
```

```python
model_sm = sm.OLS(y, X).fit()
model_sm.summary()
```

```python
model_sm.predict() # when we do not pass anything as an input the model predicts the price based on the training set
```

```python
model_sm.predict([1, 1.5]) # 1 adds a constant (not starting from the origin) and 1.5 is the carat that we would like to predict the price of
```

```python
model_sm.predict([2])
```

```python
# the model is expecting 2 columns:  one for the constant and one for th feature, 
model_sm.predict([1,2])
```

```python
# predict the entire dataframe
new_df = pd.DataFrame({'carat': [0,1,2,3,4,5]})
model_sm.predict(sm.add_constant(new_df))
```
