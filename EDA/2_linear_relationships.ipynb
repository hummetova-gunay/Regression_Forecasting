```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../Course Materials/Data/insurance.csv')
```

```python
df[['charges', 'age']].corr()
```

```python
df.corr(numeric_only=True)
```

```python
# heatmap
sns.heatmap(df.corr(numeric_only=True), annot = True, vmin=-1, vmax=1, cmap='coolwarm')
```
