# Objectives
- Build a correlation matrix heatmap
- Build a pair plot of numeric features
- Build an lmplot of 'RAM' vs. 'price'

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../Course Materials/Data/Computers.csv')
```

```python
sns.heatmap(df.corr(numeric_only=True), vmin=-1, vmax=1, cmap='coolwarm')
```

```python
sns.pairplot(df)
```

```python
sns.lmplot(df, x='ram', y='price', hue ='premium')
```
