```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Course Materials/Data/madrid_weather.csv')
```

```python
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df = df.set_index('time')
```

```python
df_monthly = df.resample('M').mean()
df_monthly.head()
```

```python
df['temperature'].plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df.columns
```

```python
hourly_weather = df[['temperature']]
```

```python
hourly_weather.plot()
```

```python
# hourly seasonality
result = seasonal_decompose(hourly_weather, period = 24)
result.plot()
```

```python
# weekly seasonality
result = seasonal_decompose(hourly_weather[:168])
result.plot()
```

```python
# Acf plot
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(hourly_weather)
```
