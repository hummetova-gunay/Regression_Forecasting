# Feature engineering
- Create 2nd and 3rd polynomials for 'hd'
- combine 'cd' and 'multi' features: if 'cd' and 'multi' equal to 'yes' then1 , 0 otherwise
- bin ram values into '<=4', '8', '16'
- create dummy variables for any categorical columns

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
```

```python
computers = pd.read_csv('../Course Materials/Data/Computers.csv')
computers.head()
```

```python
# polynomial terms
computers['hd_sq'] = computers['hd']**2
computers['hd_cube'] = computers['hd']**3
computers.head()
```

```python
# combine cd and multi
computers['ports'] = np.where((computers['cd']=='yes') & (computers['multi']=='yes'), 1, 0)
```

```python
computers.head()
```

```python
# binning

ram_dict ={
    2: '<=4', 
    4: '<=4',
    8: '8',
    16: '16'
}

computers['ram_categories'] = computers['ram'].map(ram_dict)
computers.head()
```

```python
# dummy variables

computers = pd.get_dummies(computers, drop_first=True)
computers
```
