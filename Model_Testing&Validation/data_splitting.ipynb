```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
diamonds = pd.read_csv('../Course Materials/Data/Diamonds Prices2022.csv')
```

```python
diamonds.head()
```

```python
diamonds = diamonds.sample(1000)
```

```python
X = sm.add_constant(diamonds['carat'])
y = diamonds['price']

X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=42)
print(
    f'Training Set Rows: {X.shape[0]},',
    f'Test Set Rows: {X_test.shape[0]}'
)
```

```python
insurance = pd.read_csv('../Course Materials/Data/insurance.csv')
```

```python
features =[
    'age',
    'bmi',
    'children'
]

X = sm.add_constant(insurance[features])
y = insurance['charges']

X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=42)
model = sm.OLS(y, X).fit()
model.summary()
```

```python
from sklearn.metrics import r2_score as r2
```

```python
r2(y_test, model.predict(X_test))
```

# Conceptual Reasoning behind the gap between training and test dataset: Overfitting and Underfitting

We can diagnose **overfit** and **underfit** models by comparing evaluation metrics like $R^2$, MAE and RMSE between the train and test data sets. Large gaps -> **Overfit**, poor results across both train and test scores -> **underfit**

### Solution:
**Overfit:**
- Simplify the model
- Remove features
- Regularization

**Underfit:**
- Make model more complex
- Add new features
- Feature engineering
